# Task: Top Cuisines


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('Dataset .csv')
data.head()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
# Count the numbers of each cuisine
cuisine_counts = data['Cuisines'].value_counts()

#  top three cuisines
top_three_cuisines = cuisine_counts.head(3)

# Create a table
table_data = pd.DataFrame({'Cuisine': top_three_cuisines.index,
                          })
print("Top Three Common Cuisines:")
print(table_data)
```

```python

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_three_cuisines.index, top_three_cuisines.values, color='pink')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.title('Top Three Cuisines')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Calculate percentages
total_restaurants = len(data)
percentages = (top_three_cuisines / total_restaurants) * 100
```

```python
table_data = pd.DataFrame({'Number of Restaurants': top_three_cuisines.values,
                           'Percentage of Total Restaurants': percentages.values})

print("Restaurants Details:")
print(table_data)
```

```python

# Your DataFrame with cuisine names
restaurant_details = pd.DataFrame({
    'Cuisine': ['North Indian', 'North Indian, Chinese', 'Chinese'],
    'Number of Restaurants':[936, 511, 354],
    'Percentage of Total Restaurants':[9.800021, 5.350225, 3.706418]
})

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for Number of Restaurants
ax.bar(restaurant_details.index, restaurant_details['Number of Restaurants'], color='lightblue', label='Number of Restaurants')

# Plotting the Percentage of Total Restaurants as a line plot
ax.plot(restaurant_details.index, restaurant_details['Percentage of Total Restaurants'], color='red', marker='o', label='Percentage of Total Restaurants')

# Adding labels and title
ax.set_xlabel('Cuisine')
ax.set_ylabel('Count / Percentage')
ax.set_title('Restaurant Details')

# Adding a legend
ax.legend(loc='upper right')

# Adding values on top of bars
for i, val in enumerate(restaurant_details['Number of Restaurants']):
    ax.text(i, val + 20, f'{val}', ha='center')

# Adding values on top of the percentage line plot
for i, val in enumerate(restaurant_details['Percentage of Total Restaurants']):
    ax.text(i, val, f'{val:.2f}%', ha='center', va='bottom')

plt.xticks(restaurant_details.index, restaurant_details['Cuisine'], rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
